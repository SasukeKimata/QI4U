```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_ibm_runtime import SamplerV2 as Sampler
from qiskit.visualization import plot_histogram
import numpy as np
from circle_notation_kimata import plot_qubit1, plot_qubit2, plot_qubit3, sample_and_plot_histogram
```

### 2. Circle Notationで表した状態を用意する.

1qubit状態

```python
#|0>状態
plot_qubit1(
    rQuantum0=1.0, rQuantum1=0.0,
    angle0=0, angle1=0, show_plot=True, return_state=False
)

#状態を保存する
state1_1 = plot_qubit1(
    rQuantum0=1.0, rQuantum1=0.0,
    angle0=0, angle1=0, show_plot=False, return_state=True
)
```

```python
#等確率の重ね合わせ状態①
s = 1/np.sqrt(2)
plot_qubit1(
    rQuantum0=s, rQuantum1=s,
    angle0=0, angle1=0, show_plot=True, return_state=False
)

#状態を保存する
state1_2 = plot_qubit1(
    rQuantum0=s, rQuantum1=s,
    angle0=0, angle1=0, show_plot=False, return_state=True
)
```

```python
#等確率の重ね合わせ状態② (|1>の位相を+90度動かしてみる)
s = 1/np.sqrt(2)
plot_qubit1(
    rQuantum0=s, rQuantum1=s,
    angle0=0, angle1=90, show_plot=True, return_state=False
)

#状態を保存する
state1_3 = plot_qubit1(
    rQuantum0=s, rQuantum1=s,
    angle0=0, angle1=90, show_plot=False, return_state=True
)
```

用意した1qubit状態を測定する.

```python
#|0>状態
sample_and_plot_histogram(state1_1, shots=1000, title="Sampling Result")
```

```python
#等確率の重ね合わせ状態①
sample_and_plot_histogram(state1_2, shots=10000, title="Sampling Result")
```

```python
#等確率の重ね合わせ状態②
sample_and_plot_histogram(state1_3, shots=10000, title="Sampling Result")
```

2qubit状態

```python
#等確率の重ね合わせ状態
s = 1/np.sqrt(4)
plot_qubit2(
    rQuantum00=s, rQuantum01=s, rQuantum10=s, rQuantum11=s,
    angle00=0, angle01=0, angle10=0, angle11=0,
    show_plot=True, return_state=False)

#状態を保存する
state2_1 = plot_qubit2(
    rQuantum00=s, rQuantum01=s, rQuantum10=s, rQuantum11=s,
    angle00=0, angle01=0, angle10=0, angle11=0,
    show_plot=False, return_state=True)
```

```python
#Bell状態
s = 1/np.sqrt(2)
plot_qubit2(
    rQuantum00=s, rQuantum01=0, rQuantum10=0, rQuantum11=s,
    angle00=0, angle01=0, angle10=0, angle11=0,
    show_plot=True, return_state=False)

#状態を保存する
state2_2 = plot_qubit2(
    rQuantum00=s, rQuantum01=0, rQuantum10=0, rQuantum11=s,
    angle00=0, angle01=0, angle10=0, angle11=0,
    show_plot=False, return_state=True)

```

用意した2qubit状態を測定する.

```python
#等確率の重ね合わせ状態
sample_and_plot_histogram(state2_1, shots=10000, title="Sampling Result")
```

```python
#Bell状態
sample_and_plot_histogram(state2_2, shots=10000, title="Sampling Result")
```

3qubit状態

```python
#等確率の重ね合わせ状態
s = 1/np.sqrt(8)
plot_qubit3(
    rQuantum000=s, rQuantum001=s, rQuantum010=s, rQuantum011=s,
    rQuantum100=s, rQuantum101=s, rQuantum110=s, rQuantum111=s,
    angle000=0, angle001=0, angle010=0, angle011=0,
    angle100=0, angle101=0, angle110=0, angle111=0,
    show_plot=True, return_state=False
)

#状態を保存
state3_1 = plot_qubit3(
    rQuantum000=s, rQuantum001=s, rQuantum010=s, rQuantum011=s,
    rQuantum100=s, rQuantum101=s, rQuantum110=s, rQuantum111=s,
    angle000=0, angle001=0, angle010=0, angle011=0,
    angle100=0, angle101=0, angle110=0, angle111=0,
    show_plot=False, return_state=True
)
```

```python
#GHZ状態
s = 1/np.sqrt(2)
plot_qubit3(
    rQuantum000=s, rQuantum001=0, rQuantum010=0, rQuantum011=0,
    rQuantum100=0, rQuantum101=0, rQuantum110=0, rQuantum111=s,
    angle000=0, angle001=0, angle010=0, angle011=0,
    angle100=0, angle101=0, angle110=0, angle111=0,
    show_plot=True, return_state=False
)

#状態を保存
state3_2 = plot_qubit3(
    rQuantum000=s, rQuantum001=0, rQuantum010=0, rQuantum011=0,
    rQuantum100=0, rQuantum101=0, rQuantum110=0, rQuantum111=s,
    angle000=0, angle001=0, angle010=0, angle011=0,
    angle100=0, angle101=0, angle110=0, angle111=0,
    show_plot=False, return_state=True
)
```

用意した3qubit状態を測定する.

```python
#等確率の重ね合わせ状態
sample_and_plot_histogram(state3_1, shots=10000, title="Sampling Result")
```

```python
#GHZ状態
sample_and_plot_histogram(state3_2, shots=10000, title="Sampling Result")
```
